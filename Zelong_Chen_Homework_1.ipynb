## 1biii) Selecting Training & Test Sets

```python
class_0 = v_data[v_data['class']!='AB']
class_1 = v_data[v_data['class']=='AB']

temp = class_0[0:70]
training_set = temp.append(class_1[0:140])
temp = class_0[71:]
test_set = temp.append(class_1[141:])

X_train = training_set.drop('class', axis=1)
y_train = training_set['class']

X_test = test_set.drop('class', axis=1)
y_test =test_set['class']
```

## 1ci) KNN (Euclidean metric)

```python
k_range = range(1,208,3)
scores_train = {}
scores_test = {}
scores_train_list = []
scores_test_list = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k, p=2)
    knn.fit(X_train,y_train)
    y_pred_train = knn.predict(X_train)
    y_pred_test = knn.predict(X_test)
    
    scores_train[k] = metrics.accuracy_score(y_train,y_pred_train)
    scores_train_list.append(metrics.accuracy_score(y_train,y_pred_train))
    
    scores_test[k] = metrics.accuracy_score(y_test,y_pred_test)
    scores_test_list.append(metrics.accuracy_score(y_test,y_pred_test))
```

## 1cii) Train and Test Error & Confusion Matrix, Precision, and F-score

```python
plt.plot(k_range, scores_train_list, label = 'Train')
plt.plot(k_range, scores_test_list, label = 'Test')
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy')
plt.legend(loc='upper right')
```

```python
#Find all best accuracy score of test data, and then find the one with best accuracy of train data to find K*
print("Best Accuracy=",max(scores_test_list))
# 1. KNN with Eucidean Metric
Training_Error_Rate.append(max(scores_test_list))
j = 0
temp = 0
max_occur = []
for i in scores_test_list:
    if i == max(scores_test_list):
        max_occur.append(j)
    j+=1

j=0
for i in max_occur:
    if scores_train_list[i] >= j:
        j = scores_train_list[i]
        temp = i
k_star = k_range[temp]

print("k*=",k_star)
```

```python
#Using K* to fit KNN and finding confusion matrix, precision, and F-score
#KNN with p=2 => Euclidean metric
knn = KNeighborsClassifier(n_neighbors=k_star, p=2)
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)

confusion_matrix(y_test,y_pred)
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

```python
print(classification_report(y_test,y_pred))
```

## 1ciii) Learning Curve

```python
#Inital Variables
class_0 = v_data[v_data['class']!='AB']
class_1 = v_data[v_data['class']=='AB']
N_range = range(10,210,10)
Learning_curve_scores = []

X_test = test_set.drop('class', axis=1)
y_test =test_set['class']

#Loop Start
for n in N_range:
    #Determine N and create training set
    N_div_3 = int(n/3)
    temp = class_0[0:N_div_3]
    training_set = temp.append(class_1[0:(n-N_div_3)])

    X_train = training_set.drop('class', axis=1)
    y_train = training_set['class']

    #Testing Fit with Varying values of K with increment 5 & storing accuracy scores
    #KNN with p=2 => Euclidean metric
    k_range = range(1,n,5)
    scores_train = {}
    scores_test = {}
    scores_train_list = []
    scores_test_list = []
    
    for k in k_range:
        knn = KNeighborsClassifier(n_neighbors=k, p=2)
        knn.fit(X_train,y_train)
        y_pred_test = knn.predict(X_test)

        scores_test[k] = metrics.accuracy_score(y_test,y_pred_test)
        scores_test_list.append(metrics.accuracy_score(y_test,y_pred_test))

    Learning_curve_scores.append(max(scores_test_list))

#Plot best accuracy scores of each iteration of N for Learning Curve
plt.plot(N_range, Learning_curve_scores)
plt.title('Learning Curve')
plt.xlabel('Size of Training Set N')
plt.ylabel('Testing Accuracy')
```

### B. log_10(p) = {0.1, 0.2, 0.3, ...., 1}

```python
#Inital Training and Test Data
class_0 = v_data[v_data['class']!='AB']
class_1 = v_data[v_data['class']=='AB']

temp = class_0[0:70]
training_set = temp.append(class_1[0:140])
temp = class_0[71:]
test_set = temp.append(class_1[141:])

X_train = training_set.drop('class', axis=1)
y_train = training_set['class']

X_test = test_set.drop('class', axis=1)
y_test =test_set['class']

#k* values from 1(d)iA, and log_10(p) list of values
k_star = 11
log_10_p = [x/10 for x in range(1,11,1)]
log_10_p_range = [10**(x) for x in log_10_p]
log_10_p_train_scores = []
log_10_p_test_scores = []

for p_value in log_10_p_range:
    #KNN with p=1 => Manhattan Distance
    k_range = range(1,196,5)
    scores_train = {}
    scores_test = {}
    scores_train_list = []
    scores_test_list = []
    
    knn = KNeighborsClassifier(n_neighbors=k_star, p=p_value)
    knn.fit(X_train,y_train)
    y_pred_train = knn.predict(X_train)
    y_pred_test = knn.predict(X_test)
    
    #Find accuracy score of train data
    scores_train[k] = metrics.accuracy_score(y_train,y_pred_train)
    log_10_p_train_scores.append(metrics.accuracy_score(y_train,y_pred_train))
    
    #Find accuracy score of test data
    scores_test[k] = metrics.accuracy_score(y_test,y_pred_test)
    log_10_p_test_scores.append(metrics.accuracy_score(y_test,y_pred_test))


#Find all best accuracy score of test data, and then find best accuracy of train data to find best performing Log_10_p
print("Best Accuracy=",max(log_10_p_test_scores))
# 3. KNN with Minkowski Distance B. Log_10_p value of 0.1
Training_Error_Rate.append(max(log_10_p_test_scores))
j = 0
temp = 0
max_occur = []
for i in scores_test_list:
    if i == max(scores_test_list):
        max_occur.append(j)
    j+=1

j=0
for i in max_occur:
    if log_10_p_train_scores[i] >= j:
        j = log_10_p_train_scores[i]
        temp = i
best_p_value = log_10_p_range[temp]
best_p = log_10_p[temp]
print("Best Log_10_p Value=",best_p)
```

```python
#Using Best log_10_p Value to fit KNN and finding confusion matrix
knn = KNeighborsClassifier(n_neighbors=k_star, p=best_p_value)
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)

confusion_matrix(y_test,y_pred)
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```
